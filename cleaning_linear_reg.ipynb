```python


import pandas as pd
import numpy as np

def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers



outliers_price = detect_outliers_iqr(df['price'])
print("Outliers in 'price' column:\n", outliers_price)

outliers_sqft_living = detect_outliers_iqr(df['sqft_living'])
print("\nOutliers in 'sqft_living' column:\n", outliers_sqft_living)



print("\nNumber of outliers in 'price':", len(outliers_price))
print("Number of outliers in 'sqft_living':", len(outliers_sqft_living))


numerical_cols = ['price', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long']

for col in numerical_cols:
    outliers = detect_outliers_iqr(df[col])
    print(f"\nOutliers in '{col}' column:\n{outliers}")
    print(f"Number of outliers in '{col}': {len(outliers)}")
```
